```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Use a raw string for the file path
image_path = r'C:/Users/panca/OneDrive/Documents/praktikum_pcd/modul4/manchester_united.jpeg'
image = cv2.imread(image_path) 

# Convert the image to HSV
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) 

# Convert the image back to BGR and then to grayscale
image2 = cv2.cvtColor(image, cv2.COLOR_HSV2BGR)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# Set the dimension of the kernel
figure_size = 9 

# Apply the mean filter
new_image = cv2.blur(image2,(figure_size, figure_size))

# Display the original and filtered images
plt.figure(figsize=(11,6))
plt.subplot(121), plt.imshow(image2, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(new_image, cmap='gray'),plt.title('Mean filter')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Apply the median filter
new_image = cv2.medianBlur(image2, figure_size)

# Display the original and filtered images
plt.figure(figsize=(11,6))
plt.subplot(121), plt.imshow(image2, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(new_image, cmap='gray'),plt.title('Median Filter')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import numpy as np
import cv2

def mode(arr):
    from scipy import stats
    return stats.mode(arr)[0][0]

def apply_mode_filter(gr, adj):
    output = gr.copy()
    size = (2 * adj + 1) ** 2
    for i in range(adj, gr.shape[0] - adj):
        for j in range(adj, gr.shape[1] - adj):
            arr1 = gr[i-adj:i+adj+1, j-adj:j+adj+1].flatten()
            get_mode = mode(arr1)
            output[i, j] = get_mode
    return output

# Example usage:
gr = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)
adj = 1
output = apply_mode_filter(gr, adj)
cv2.imshow('Output', output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
dft = cv2.dft(np.float32(image2),flags = cv2.DFT_COMPLEX_OUTPUT)

# shift the zero-frequncy component to the center of the spectrum
dft_shift = np.fft.fftshift(dft)

# save image of the image in the fourier domain.
magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1]))

# plot both images
plt.figure(figsize=(11,6))
plt.subplot(121),plt.imshow(image2, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```
